```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def entropy(px, base):
    """
    px : array like
    """
    if base=="e":
        log = np.log
    elif base=="2":
        log = np.log2
    elif base=="10":
        log = np.log10
    
    px = np.array(px).astype(float)
    px[px==0] = 1e-10 #log(0)は計算できないため、小さな値に置き換えておく
    return -1 * np.sum(px*log(px))

p_list = np.linspace(0, 1, 100)
ent_e =  np.zeros(len(p_list))
ent_2 =  np.zeros(len(p_list))
ent_10 =  np.zeros(len(p_list))

for i, p in enumerate(p_list):
    ent_e[i] = entropy(np.array([p, 1 - p]), base="e")
    ent_2[i] = entropy(np.array([p, 1 - p]), base="2")
    ent_10[i] = entropy(np.array([p, 1 - p]), base="10")    
    
plt.plot(p_list, ent_e, label='entropy_e')
plt.plot(p_list, ent_2, label='entropy_2')
plt.plot(p_list, ent_10, label='entropy_10')

plt.grid(which='major', color='black', linestyle=':')
plt.grid(which='minor', color='black', linestyle=':')
plt.legend(loc='best')
plt.xlabel("p(x)")
plt.show()
```
