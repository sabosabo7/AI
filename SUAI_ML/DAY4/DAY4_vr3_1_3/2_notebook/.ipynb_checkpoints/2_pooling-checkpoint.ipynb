# プーリングをやってみる

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from common.util import im2col
```

## 簡単なデータでプーリングをやってみる

### 入力データをつくる

```python
img = np.array([[[[1,2,1,0],[0,1,2,3],[3,0,1,2],[2,4,0,1]]]])
img
```

### プーリング演算用に配列を変形する
一般的にプーリングでは、ウィンドウサイズとストライドは同じにする

```python
col = im2col(img,2,2,stride=2,pad=0)
print(col.shape)
col
```

### 最大値プーリングの実行

```python
out = np.max(col,axis=1)
out
```

```python
img_y = out.reshape(1,2,2,1).transpose(0,3,1,2)
img_y
```

## MINIST画像でプーリングをやってみる

```python
# Load the MNIST dataset
import tensorflow as tf
mnist = tf.keras.datasets.mnist
(X_train, y_train),(X_test, y_test) = mnist.load_data()

def show_image(img): 
    pil_img = Image.fromarray(img)
    plt.imshow(img)
    plt.gray()
    plt.show()
    return
```

### trainの1枚目を取り出す

```python
label = y_train[0]
img = X_train[0]
print("label=%s"%label)
img = img.reshape(28,28)
img = np.uint8(img*255)
show_image(img)
```

### プーリング演算用に配列を変換する


```python
img = img.reshape(1,1,28,28)
col = im2col(img,3,3,stride=1,pad=0)
print(col.shape)
col
```

## 最大値プーリングの実行

```python
out = np.max(col,axis=1)
out
```

```python
img_y = out.reshape(26,26)
img_y
```

```python
show_image(img_y)
```

## 他の写真でプーリングを試してみる

```python
img_ = Image.open("../1_data/bridge.jpg")
img_
```

```python
img_ =  np.asarray(img_)
print("次元:",img_.shape)
img_
```

```python
img_ = img_.transpose(2,0,1)
print("次元:",img_.shape)
img_
```

### プーリング用に配列を変換する

```python
col_ = im2col(img_.reshape(1,3,224,224),4,4,stride=4,pad=0)
print(col_.shape)
col_
```

### チャンネル方向のデータが横に並んでいるので、縦に並べ替える

```python
col_ = col_.reshape(-1,4*4)
print(col_.shape)
col_
```

### 最大値プーリングの実行

```python
out_ = np.max(col_,axis=1)
print(out_.shape)
out_
```

```python
img_y = out_.reshape(56,56,3)
print(img_y.shape)
show_image(np.uint8(img_y))
```

## ウィンドウサイズを変えてみる

```python
for size in [4,8,16,32]:
    print("size=%s"%size)
    col_ = im2col(img_.reshape(1,3,224,224),size,size,stride=size,pad=0)
    col_ = col_.reshape(-1,size*size)
    out_ = np.max(col_,axis=1)
    img_y = out_.reshape(int(224/size),int(224/size),3)
    show_image(np.uint8(img_y))
```
